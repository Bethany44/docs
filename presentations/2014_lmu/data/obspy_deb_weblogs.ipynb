```python
from datetime import datetime

def date_parser(value1, value2):
    """
    Ignore the timezone...
    """
    return datetime.strptime(value1,
        "[%d/%b/%Y:%H:%M:%S")
```

```python
# Simple way to read a whole directory full of logs into a pandas data
# frame. The date parsing is what makes it really slow...maybe disable it?

from datetime import datetime
import glob
import pandas as pd

# Get the files
files = glob.glob("/home/megies/temp/obspy_lunchtime_talk/obspy-deb.access.2014-06-22.0.9.2.unique.one-spaced.summary")

all_logs = None

for filename in files:
    # Some files are gzip compressed. Pandas can also directly read those files.
    if filename.endswith("gz"):
        compression="gzip"
    else:
        compression=None
        
    df = pd.io.parsers.read_csv(filename, sep=" ", header=None, quotechar='"',
        #names=["host", "ip", "", "", "time1", "time2", "full_request", "", "", "", "user_agent"],
        #names=["ip", "", "", "time1", "time2", "", "request", "", "", "", "", "user_agent"],
        names=["ip", "", "", "time1", "time2", "full_request", "", "", "", "user_agent"],
        #parse_dates={"time": [4, 5]}, date_parser=date_parser, compression=compression, escapechar="\\")
        parse_dates={"time": [3, 4]}, date_parser=date_parser, compression=compression, escapechar="\\")
    # Remove all empty columns.
    del df[""]
    # Split the full request into the actual request URI and the HTTP method.
    # XXX: Is there a better method?
    df["request"] = df.apply(lambda row: row["full_request"].split()[1] if " " in row["full_request"] else "", axis=1)
    df["method"] = df.apply(lambda row: row["full_request"].split()[0], axis=1)
    del df["full_request"]
    
    if all_logs is not None:
        all_logs = all_logs.append(df)
    else:
        all_logs = df
```

```python
# Get the number of downloads for the debian packages for 0.9.2
debs_0_9_2 = all_logs[all_logs["request"].str.contains("/pool/main/p/python-obspy/python-obspy_0.9.2")]
# Get the older ones as well. The old ones will have multiple packages per python installation.
#debs_old = all_logs[all_logs["request"].str.contains("/pool/main/p/python-obspy-")]
#debs_old = debs_old[debs_old["request"].str.contains(".deb")]
print "Downloads Debian Packages 0.9.2:", len(debs_0_9_2)
#print "Downloads Older Debian Packages:", len(debs_old)
```

```python
# Try to get a list of all ip addresses

#ips = ObsPy_OSX_0_8_3["ip"]
#ips = ips.append(ObsPy_OSX_0_7_x["ip"])
#ips = ips.append(ObsPy_OSX_Rest["ip"])
#ips = ips.append(debs_0_8_3["ip"])
#ips = ips.append(debs_old["ip"])

ips = debs_0_9_2["ip"]

# Create unique list.
ips = list(set(list(ips)))

print "%i unique ips found" % len(ips)
```

```python
# Now attempt to geocode the whole thing.
import json
import requests
import time

successful_lookups = []
failed_lookups = []

for ip in ips:
    print ".",
    # Try up to ten times for each.
    success = False
    for _i in range(10):
        try:
            lookup = json.loads(requests.get("http://freegeoip.net/json/%s" % ip).content)
            successful_lookups.append(lookup)
            success = True
            break
        except:
            time.sleep(0.1)
            continue
    if success is False:
        failed_lookups.append(ip)
        

print "\n"
print "Geocoded %i ips, %i failed to geocode" % (len(successful_lookups), len(failed_lookups))
```

```python
# Convert to pandas data frame again.
ip_data = pd.DataFrame.from_records(successful_lookups)
# pickle!
import pickle
pickle.dump(ip_data, open("/home/megies/temp/obspy_lunchtime_talk/pandas.pkl", "wb"))
ip_data[0:10]
```
