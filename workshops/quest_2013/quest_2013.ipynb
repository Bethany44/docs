### 1. Python in nutshell

Very quick overview for people with knowledge in other programming languages.



#### 1.1. Variables

Variables are assigned with the equal sign - no need to specify the type (dynamically typed language).

```python
a = "a"
b = 1
c = 1.3
d = True
e = False

print type(a)
print type(b)
print type(c)
print type(d), type(e)
```

#### 1.2. Container types

The most used container types are lists and dictionaries.

Lists are an ordered collection of objects. Everything in Python is an object so it can contain numbers, strings, other lists, functions, ...

They are created with squared brackets.

```python
# Create with square brackets.
some_list = [1, "a", [1, 2], True]
# Add a new elements with the append() method.
some_list.append(False)
print some_list
print some_list[0]

# Python uses zero-based indexing. Using a negative index will cause Python to add
# the length of the list to the index. Thus one can access elements at the back.
print some_list[0], some_list[-1]
```

Dictionaries store arbitrary key, value pairs. The difference to a list is that one does not use an index to refer to a certain element in it but rather a key. In other languages this is also known as a hashmap or an associative array.

```python
some_dict = {"a": 1, "b": 2}
some_dict["c"] = 3
print some_dict
print some_dict["b"]
```

#### 1.3 Control flow

Intendation groups code blocks! No need for brackets.

```python
a = 1
b = 2

if a == 1 and b != 1:
    print "a is 1 and b not"
elif b == 1:
    print "b is not 1"
else:
    print "Something else happened"
```

```python
# Python support "for each" loops. One essentially
# iterates over every item in a container.
some_list = ["a", "b", "c"]
for i in some_list:
    print i
    
# To loop a certain amount of time, use the xrange() function.
sum = 0
for i in xrange(10):
    sum += i
print sum
```

### 2. Importing ObsPy

The first step is to import ObsPy. The top-level namespace contains ObsPy's most used functions and objects:

* The **read()** function. Reads all kinds of waveform formats.
* The **readEvents()** function. Reads QuakeML files.
* The **UTCDateTime** class. Used to deal with times.
* The **Trace** class. Deals with a single chund of continuous waveform data.
* The **Stream** class. A collection of traces. Eases organization and multi-component processing.

You always have to import any functionality that is not part of the actual Python language.

```python
import obspy

# Print the contents of the root ObsPy module discarding currently
# not needed functionality.
print [_i for _i in dir(obspy) if not _i.startswith("_") and _i != "core"]
```

### 3. Dealing with Time
All times within ObsPy are handled with the **UTCDateTime** class.

* Flexible way of creating time objects.
* Time differences are always **expressed in seconds**.

```python
time_of_arrival = obspy.UTCDateTime(2013, 5, 19, 18)
wakeup_time = obspy.UTCDateTime("2013-05-21T07:30:00")
right_now = obspy.UTCDateTime()

print "Awake for", right_now - wakeup_time, "seconds."
```

### 4. Waveform data

Read some files.

```python
st = obspy.read("./event.CI.PHL.LHZ.1998.196.1998.196.mseed")
st.filter("bandpass", freqmin=0.1, freqmax=0.2)
st.trim(st[0].stats.starttime + 60 * 50, st[0].stats.endtime - 40 * 60)
st.resample(10.0)
print st
st.plot()

for tr in st:
    print tr.data.ptp()
```
